## Normal distribution, histogram (density function)

```python
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt

mu, sigma = 100, 15 # mean (like IQ), standard deviation
x = mu + sigma*np.random.randn(10000) # generates elements of mu mean and mu std
									  # numpy.random.randn -> 0 mean, 1 std
									  
									  
n, bins, patches = plt.hist(x, 50, normed=1, facecolor='green', alpha=0.75)
				# https://matplotlib.org/devdocs/api/_as_gen/matplotlib.pyplot.hist.html
				# Creates a histogram structure (tuple)
				# x -> input data
				# 50 -> Number of bins: number defined + 1 
				# normed=1 -> modifies y axis, if 1, the area under the curve will be 1. Normalization: the occurrence density of the case divided by all the cases 
				# facecolor='green' -> color of the bins
				# alpha=0.75 -> gives the shade of color in light
				
# add a 'best fit' line
y = mlab.normpdf( bins, mu, sigma)
l = plt.plot(bins, y, 'r--', linewidth=1)	

plt.xlabel('IQ')
plt.ylabel('Probability')
plt.title(r'$\mathrm{Histogram\ of\ IQ:}\ \mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)

plt.show()
```

## Playing with your own data

```python
x = sorted([186, 176, 158, 180, 186, 168, 168, 164, 178, 170, 189, 195, 172,
     187, 180, 186, 185, 168, 179, 178, 183, 179, 170, 175, 186, 159,
     161, 178, 175, 185, 175, 162, 173, 172, 177, 175, 172, 177, 180]) 
	
n, bins, patches = plt.hist(x, bins= 10, normed=1, facecolor='blue', alpha=0.75)
        # bins = 'auto'
		# histtype='step'

y = mlab.normpdf( bins, np.mean(x), np.std(x))
l = plt.plot(bins, y, 'r--', linewidth=2)	


plt.xlabel('Random')
plt.ylabel('Propability')
plt.title(r'Hisztogram')

#plt.grid(True)

plt.show()
```

## Pie chart
### The 1994 election of Hungary

```python

labels = 'Magyar Szocialista Párt (MSZP)', 'Szabad Demokraták Szövetsége (SZDSZ)', 'Magyar Demokrata Fórum (MDF)', 'Független Kisgazdapárt (FKGP)', 'Kereszténydemokrata Néppárt (KDNP)', 'Fiatal Demokraták Szövetsége (Fidesz)', 'Egyéb'
sizes = [31.27, 18.62, 12.03, 7.88, 7.37, 7.70, 15.13]
explode = (0, 0, 0, 0, 0, 0, 0.1)  

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
