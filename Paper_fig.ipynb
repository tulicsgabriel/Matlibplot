# Actual fig from a paper

### Classification accuracies in function of feature numbers using SVM-rbf classification method for merged (solid line) and separate (dashed line) MA, FD and RP classes.

Paper: Automatic discrimination of several types of pathological speech disorders

```python
import matplotlib.pyplot as plt
import pandas as pd

# Assign spreadsheet filename to `file`
file = 'paper_data.xlsx'

# Load spreadsheet
xl = pd.ExcelFile(file)

# Print the sheet names
print(xl.sheet_names)

# Load a sheet into a DataFrame by name: df1
dataset = xl.parse('data')

X=dataset.loc[:,'Index']
y1=dataset.loc[:,'Merged OD, FD and RP classes']
y2=dataset.loc[:,'Separate OD, FD and RP classes']
y3=dataset.loc[:,'Separate OD, FD and RP classes - female']
y4=dataset.loc[:,'Separate OD, FD and RP classes - male']
y5=dataset.loc[:,'Merged OD, FD and RP classes -female']
y6=dataset.loc[:,'Merged OD, FD and RP classes - male']

plt.plot(X, y1*100, marker = 'p',markersize=10, linestyle='-',  linewidth=2, label='Merged MA, FD and RP classes') #  marker = 'o',markersize=10, linestyle='--',  linewidth=4,
plt.plot(X, y5*100, marker = 'h',markersize=10, linestyle='-.',  linewidth=2, label='Merged MA, FD and RP classes -female')
plt.plot(X, y6*100, marker = 'x',markersize=10, linestyle='--',  linewidth=2, label='Merged MA, FD and RP classes - male')
plt.plot(X, y2*100, marker = '*',markersize=10, linestyle='--',  linewidth=2, label='Separate MA, FD and RP classes')
plt.plot(X, y3*100, marker = '+',markersize=10, linestyle=':',  linewidth=2, label='Separate MA, FD and RP classes - female')
plt.plot(X, y4*100, marker = '^',markersize=10, linestyle='-',  linewidth=2, label='Separate MA, FD and RP classes - male')

plt.xlabel('Number of features', fontsize =20)
plt.ylabel('Classification accuracy [%]', fontsize = 20)
plt.legend(bbox_to_anchor=(1.05, 1), loc='best', prop={'size': 14}, borderaxespad=0.)
plt.grid(True)
  # t nem szabad lehagyni!  jeleti meg a label-t
			  # loc = 'upper right' 
			  # loc = 'best'
			  # https://matplotlib.org/api/legend_api.html
#fig = plt.gcf()
#fig.set_size_inches(18.5, 10.5)
#fig.savefig('test2png.png', dpi=500)
plt.show()
```
